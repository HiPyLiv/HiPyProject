Decision Trees are a set of very unique models in Python - instead of outputting an equation or complicated algorithm, the decision tree method will output a series of questions that result in data categorisation, which we can either tell the computer to follow, or follow ourselves. The major upside of the decision tree method is that our results are extremely easy to understand and can be can visualised simply - however in general, decision trees do a worse job at categorising data than other methods due to their simplicity.

## Using Python to Create Decision Trees

The algorithm for creating a decision tree is more complicated than the other machine learning methods, so we won't be going over it here - the intuition behind it is categorising the data by the variable which splits the data the best first, then continuing down until we reach a predefined depth. Let's get right into some code to see how this works in practice:

First we need to import the modules we'll need to perform our analysis:

```python
from sklearn.datasets import load_iris
from sklearn import tree
import numpy as np
import pandas as pd
```

Then let's import some data from a scikitlearn dataset - the iris dataset:

```python
iris = load_iris()
```

This data isn't stored as a pandas dataframe, instead it's stored as numpy arrays under different methods of the iris object. For example, iris.feature_names and iris.data give us the feature names and features of our data, and iris.target_names and iris.target give us the category names and categories of our data.

```python
print(iris.feature_names)
print(iris.data[:5])
print(iris.target_names)
print(iris.target[:5])
```

Creating a decision tree is the easy bit - as with all the other machine learning algorithms we can just call the function and fit the data:

```python
clf = tree.DecisionTreeClassifier().fit(iris.data, iris.target)
```

From here we can predict what class data points will end up in, as usual, using the predict method:

```python
print( clf.predict([[1,2,3,4],[8,7,6,5]]) )
```

This for the most part is enough - we can take some new points in and use the model to predict them. However the advantage of using the decision tree method over other classification methods is that we can visualise the tree with a bit of work.

## Visualising a Decision Tree

The great thing about using a decision tree is that it can be visualised to an easy format. However, doing this requires us to jump through some very annoying hoops. First off, you will need to install Graphviz from [here](http://www.graphviz.org/Download..php), then if you're on Windows, add the Graphviz bin folder to your PATH environment variables. Finally, you'll need to install the pydotplus and graphviz libraries using Anaconda Navigator. Graphviz is a standalone visualisation software, so this won't work on Azure Notebooks at all.

Luckily, that's the hardest part - all we need to do now is run the following code and we're good to go! We start by importing all our libraries, create the graphviz object, turn it into a dot object, then create a png from that.


```python
import pydotplus 
from IPython.display import Image
import graphviz
dot_data = tree.export_graphviz(clf, out_file=None, feature_names=iris.feature_names,class_names=iris.target_names)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png()) 
```

Now we can make some predictions ourself!
